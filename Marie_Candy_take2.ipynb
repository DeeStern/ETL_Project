```python
#Import dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
#Read csv
csv_file = "halloween_candy.csv"
halloween_candy_df = pd.read_csv(csv_file, sep=";")
halloween_candy_df.head()
```
